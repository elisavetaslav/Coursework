### Библиотеки

**Библиотеки** в Python - это наборы готовых функций и инструментов, которые другие люди написали и поделились с сообществом. Они обычно создаются для решения конкретных задач или предоставления определенного функционала, который может быть использован в ваших программах.

Когда вы используете библиотеку в своем коде, вы можете просто импортировать ее, чтобы получить доступ к ее функциям и классам. Библиотеки позволяют вам не писать все с нуля, а вместо этого использовать готовые решения для решения задач быстрее и эффективнее. 

Представьте, что вы хотите сделать что-то сложное, например, построить дом. Вам нужно использовать разные инструменты, такие как молоток, пила, гвозди и т. д. Но вы не создаете эти инструменты с нуля.

Библиотеки в Python - это как ящики с инструментами для программирования. Когда вам нужен какой-то инструмент, вы просто берете его из библиотеки и используете.

#### Как подключить библиотеку?

В Python большое количество библиотек входят в стандартную библиотеку. Она доступна всем по умолчанию. И чтобы использовать необходимо просто импортировать нужную.

Примеры таких библиотек:
+ os - для работы с операционной системой
+ sys - для взимодействия с интерпритатором
+ math - для работы с математическими функциями
+ datetime - для работы с датой и временем
+ random - для генерации случайных чисел
и т.д

Но что же делать с другими библиотеками, не со стандартными?

Для начала надо скачать ее на компьютер.
В Python есть менеджер пакетов, например pip или conda.

1. Открываем терминал или командную строку
    (В PyCharm правая кнока мыши -> Open in terminal)
2. Пишем `pip install name_library`
    (Если есть проблема с pip, проверьте правильную установку питона)

Как импортировать библиотеку?
Все просто
`import name_library`

`from name_library import method_from_library` импортировать только один метод

`import name_library as lib` (какое-нибудь сокращение) импортировать библиотеку и называть по-другому

Чтобы вызвать метод из библиотеки нужно написать название библиотеки и через точку позвать нужный метод. `name_library.needed_method()`

### Math

Модуль math предоставляет математические функции и константы для работы с числами. Он включает в себя такие функции, как тригонометрические, экспоненциальные и логарифмические функции, округление чисел, вычисление факториала и биномиального коэффициента, а также другие математические операции. Этот модуль широко используется в научных вычислениях, инженерных расчетах, статистике и других областях, где требуется выполнение математических операций.


```python
import math

x = 25
math.sqrt(x) # вычисление корня
```

```python
import math as mth

angle = mth.radians(40) # задаем угол из градусов в радианы
(mth.sin(angle), mth.cos(angle)) # вычисляем косинус и синус
```

```python
print(f'e^2 = {math.exp(3)**2}') # экспонента
print(f'ln{x} = {math.log(x)}') # натуральный логарифм
print(f'The pi is {math.pi}') # число пи
print(math.inf) # Плюс бесконечность = float(inf)

print(math.ceil(24.5)) # округление вверх
print(math.floor(24.5)) # округление вниз
print(int(24.5)) # математическое округление для сравнения

print(math.comb(4, 2)) # биномиальный коэффициент
print(math.factorial(4)) # факториал
lst = [1.234355656, 2.234546999, 3000000.234]
print(math.fsum(lst)) # точная сумма для float без потери точности
print(math.gcd(24, 120, 8)) # наибольшее общее значение
print(math.trunc(24.5), math.trunc(-24.5)) # удаление дробной части, floor : x > 0, ceil : x < 0
```

### Random

Модуль random предоставляет функции для работы с случайными числами. Она позволяет генерировать случайные числа, выбирать случайные элементы из последовательностей, перемешивать элементы и многое другое. Благодаря этой библиотеке можно создавать случайные данные для тестирования программ, моделирования случайных процессов или принятия случайных решений в алгоритмах и приложениях.

```python
import random


print(random.random()) # генерирует случайное число в диапазоне от 0 до 1
print(random.randint(1, 100)) # генерирует случайное целое число в диапазоне
print(random.uniform(1, 10)) # генерирует случайное число с плавающей точкой в диапазоне
print(random.randint(1, 100), 5) # генерирует случайное целое число в диапазоне c шагом


# Выбирает случайный элемент из списка
my_list = ['one', 'two', 'three', 'four', 'five']
print(random.choice(my_list))

# Перемешивает элементы списка
random.shuffle(my_list)
print(my_list)

# Генерирует случайный выбор из списка с заданными вероятностями
choices = ['rock', 'paper', 'scissors']
weights = [0.3, 0.3, 0.4]  # вероятности rock, paper, scissors
rand_choice = random.choices(choices, weights=weights, k=2)
print(rand_choice)
```

### Datetime, time

Модуль **datetime** предоставляет классы и функции для работы с датами и временем. Он включает классы datetime, date и time, которые представляют комбинации даты и времени, даты и только времени соответственно.

Модуль **time** предоставляет функции для работы с временем в виде числовых значений, основанных на эпохе Unix (отсчитывает время с 1 января 1970 года).
 Он включает функции для засечения времени выполнения кода, ожидания определенного интервала времени, измерения системного времени и т. д.

```python
import datetime

print(datetime.datetime.now()) # текущая дата и время

print(datetime.date.today()) # текущая дата

custom_date = datetime.date(2022, 4, 26)
print(custom_date) # создание объекта даты
print("Day:", custom_date.day, "Month:", custom_date.month, "Year:", custom_date.year)

custom_time = datetime.time(13, 45, 30)
print(custom_time) # создание объекта времени
print("Hour:", custom_time.hour, "Minute:", custom_time.minute, "Second:", custom_time.second)

# Форматирование даты и времени в строку
print(datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"))
# И форматирование обратно
print(datetime.datetime.strptime("2022-04-15 13:45:30", "%Y-%m-%d %H:%M:%S"))
```

### NumPy

Библиотека **numpy** позволяет выполнять высокоуровневые математические функции с многомерными массивы и делать это максимально быстро и эффективно.

Одним из ключевых преимуществ **numpy** - многомерные массивы. **ndarray**  позволяет эффективно хранить и оперировать многомерными данными, такими как изображения, звуки, временные ряды и т.д.

Также **NumPy** обеспечивает эффективные функции для выполнения математических операций над массивами, такие как умножение, сложение, деление, вычитание и т. д. Благодаря своей внутренней реализации на языке C, операции в NumPy выполняются гораздо быстрее, чем обычные операции в Python.

Давайте проведем небольшой эксперимент и убедимся в эффективности numpy. 
Перемножим две большие матрицы и сравним время.

```python
import numpy as np
import time

# Создадим две матрицы размером 100x100 используя обычные списки
matrix1 = [[i+j for i in range(100)] for j in range(100)]
matrix2 = [[i-j for i in range(100)] for j in range(100)]

#и использую array из numpy
mtx1= np.array(matrix1)
mtx2= np.array(matrix2)


start_time = time.time()
result_list = [[0 for _ in range(100)] for _ in range(100)]
for i in range(100):
    for j in range(100):
        for k in range(100):
            result_list[i][j] += matrix1[i][k] * matrix2[k][j]
end_time = time.time()
res_time = end_time - start_time
print("Время для списков Python:", end_time - start_time)


start_time = time.time()
result_array = np.dot(mtx1, mtx2) # перемножаем две матрицы
end_time = time.time()
print("Время для массивов NumPy:", end_time - start_time)

print(f"В данной задаче Numpy быстрее в {int(res_time/(end_time - start_time))} раз.")
```

Конечно тут мы использовали абсолютно не эффективный алгоритм, попробуем на более простой задаче: Сложить два списка.

```python
import numpy as np
import time

n = 1000000
start_time = time.time()
list1 = [i for i in range(n)]
list2 = [i for i in range(n)]
result_list = [list1[i] + list2[i] for i in range(len(list1))]
end_time = time.time()
res_time = end_time - start_time
print("Время для списков Python:", end_time - start_time)


start_time = time.time()
array1 = np.arange(n)
array2 = np.arange(n)
result_array = array1 + array2
end_time = time.time()
print("Время выполнения для массивов NumPy:", end_time - start_time)
print(f"В данной задаче Numpy быстрее в {int(res_time/(end_time - start_time))} раз.")
```

Это уже более реально число, и правда NumPy примерно в 30 раз быстрее, но все конечно зависит от задачи.

Что ж, когда мы удостоверились в эффективности этой библиотеки, давайте подробнее рассмотрим ее методы.

```python
#Разберемся с основой как создать NumPy array
arr = np.array([1, 2, 3, 4])
print(arr, end='\n\n')

arr = np.array([[1, 2], [3, 4]])
print(arr, end='\n\n')

arr = np.array([[[1],[2]],[[3], [4]]])
print(arr, end='\n\n')
```

```python
# numpy.arange(): Создает массив с последовательными числами.
arr = np.arange(1, 10)
print(arr, end='\n\n')

# numpy.zeros(): Создает массив из нулей.
arr = np.zeros(2)
print(arr, end='\n\n')

# numpy.ones(): Создает массив из единиц.
arr = np.ones((2, 4))
print(arr, end='\n\n')

# numpy.empty(): Создает массив неинициализированных значений.
arr = np.empty((3, 2))
print(arr, end='\n\n')
```

```python
# ndarray.shape: Возвращает размеры массива.
print(arr.shape)

# ndarray.dtype: Возвращает тип данных элементов массива.
print(arr.dtype)

# ndarray.reshape(): Изменяет форму массива.
arr = np.array([[1, 2], [3, 4], [5, 6]])
print(arr)
print(np.reshape(arr, (2, 3)), end='\n\n')
arr = np.reshape(arr, (2, 3), order='F')
print(arr, end='\n\n')

# ndarray.flatten(): Преобразует многомерный массив в одномерный.
print(arr.flatten(), end='\n\n')

# numpy.concatenate(): Объединяет массивы.
arr = np.ones((2, 4))
arr2 = np.zeros((2, 4))
print(np.concatenate((arr, arr2)), end='\n\n')
print(np.concatenate((arr, arr2), axis=1), end='\n\n')
```

```python
# numpy.add(): Сложение элементов.
arr = np.array([[1, 2], [3, 4], [5, 6]])
arr2 = np.array([[1, 0], [0, 1], [1, 2]])
print(np.add(arr, arr2), end='\n\n')
print(arr + arr2, end='\n\n')

# numpy.subtract(): Вычитание элементов.
print(np.subtract(arr, arr2), end='\n\n')
print(arr - arr2, end='\n\n')

# numpy.multiply(): Поэлементное умножение.
print(np.multiply(arr, arr2), end='\n\n')
print(arr * arr2, end='\n\n')

# numpy.divide(): Поэлементное деление.
print(np.divide(arr2, arr), end='\n\n')
print(arr2 / arr, end='\n\n')

# numpy.dot(): Скалярное произведение или умножение матриц.
arr = np.array([[1, 2], [3, 4], [5, 6]])
arr2 = arr.T
print(arr2, end='\n\n')
print(arr.dot(arr2), end='\n\n')
print(arr.flatten().dot(arr2.flatten()), end='\n\n')
```

```python
# numpy.sum(): Сумма элементов массива.
print(np.sum(arr))

# numpy.mean(): Среднее значение элементов массива.
print(np.mean(arr))

# numpy.max(): Максимальное значение элементов массива.
print(np.max(arr))

# numpy.min(): Минимальное значение элементов массива.
print(np.min(arr))

# numpy.std(): Стандартное отклонение элементов массива.
print(np.std(arr), end='\n\n')

# numpy.exp(): Экспонента каждого элемента массива.
print(np.exp(arr), end='\n\n')

# numpy.log(): Натуральный логарифм каждого элемента 
print(np.log(arr), end='\n\n')

print(np.sin(arr), end='\n\n')
```

```python
# numpy.random.rand(): Случайные числа из равномерного распределения.
print(np.random.rand())

# numpy.random.randn(): Случайные числа из нормального распределения.
print(np.random.randn())

# numpy.random.randint(): Случайные целые числа в заданном диапазоне.
print(np.random.randint(1, 10))
```

```python
# Посмотрим, что же это за рандомные числа, и построим графики
arr1 = np.random.rand(50)  # равномерное распределение
arr2 = np.random.randn(50)   # нормальное распределение
arr3 = np.random.randint(-10, 10, size=50)  # случайные целые числа в диапазоне [-10, 10)

plt.figure(figsize=(10, 5))

plt.plot(arr1, label='Равномерное распределение')
plt.plot(arr2, label='Нормальное распределение')
plt.plot(arr3, label='Случайные числа')

plt.title('Три графика с разными случайными числами')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(arr1)
plt.title('Равномерное распределение')

plt.subplot(1, 3, 2)
plt.plot(arr2)
plt.title('Нормальное распределение')

plt.subplot(1, 3, 3)
plt.plot(arr3)
plt.title('Случайные числа')

plt.show()
```

На самом деле функционал этой библиотеки огромен и применяется в различных областях научныхх вычислений, а также в инженерии, анализе данных и машинном обучении, в финансовой аналитике, в обработке изображений и видео.
